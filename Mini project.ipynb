```python
# membuat list kosong untuk data kolom choice
data_choice = {'choice': []}
```

```python
# mengambil data untuk list choice 
# Iterasi melalui setiap baris di df
for _, row in df.iterrows():
    user_num = row['User_phone']
    
    # Iterasi melalui setiap kolom pertanyaan di df
    for kolom in df.columns[2:]:
        pertanyaan = row[kolom].split(', ')
        
        # Iterasi melalui setiap pilihan dalam pertanyaan
        for choice in ['A', 'B', 'C']:
            choice_value = 1 if choice in pertanyaan else 0
            
            data_choice['choice'].append(choice_value)
```
